Data Cleaning: Handling missing values in text fields

```python
df["overview"] = df["overview"].fillna("")
df["tagline"] = df["tagline"].fillna("")
df["keywords"] = df["keywords"].fillna("")
```

Combining relevant metadata into a single searchable document text block

```python
df["searchable_text"] = (
    df["title"] + " " + df["tagline"] + " " + df["overview"] + " " + df["keywords"]
)

print(f"Dataset loaded. Total movies to index: {len(df)}")
```

Preprocessing Function

```python
def preprocess_text(text):
    """Tokenizes text, removes punctuation, and converts to lowercase."""
    if not isinstance(text, str):
        return []
    text = text.lower()
    tokens = re.findall(r"\b\w+\b", text)
    return tokens

sample_text = "Iron Man 2: Fast-paced action, Stark's tech!"
print("Original:", sample_text)
print("Tokens:  ", preprocess_text(sample_text))
```

Building the Inverted Index

```python
inverted_index = defaultdict(dict)

for _, row in df.iterrows():
    doc_id = int(row["id"])
    tokens = preprocess_text(row["searchable_text"])

    for token in tokens:
        inverted_index[token][doc_id] = inverted_index[token].get(doc_id, 0) + 1

print(f"Inverted Index successfully built.")
print(f"Total unique terms indexed: {len(inverted_index)}")

sample_word = "space"
if sample_word in inverted_index:
    print(
        f"\nThe word '{sample_word}' appears in {len(inverted_index[sample_word])} movies."
    )
```

TF-IDF Vector Normalization

```python
doc_lengths = {}
N = len(df)

for _, row in df.iterrows():
    doc_id = int(row["id"])
    length_sq = 0

    for term in inverted_index:
        if doc_id in inverted_index[term]:
            tf = inverted_index[term][doc_id]
            df_t = len(inverted_index[term]) 

            idf = math.log(N / (1 + df_t))

            length_sq += (tf * idf) ** 2

    doc_lengths[doc_id] = math.sqrt(length_sq) if length_sq > 0 else 1.0

print("Document vector lengths successfully computed for ranking optimization.")
```

Boolean Retrieval Engine (AND / OR Operations)

```python
def boolean_retrieval(query, operator="OR"):
    """Finds the initial set of candidate document IDs matching the query constraints."""
    terms = preprocess_text(query)
    if not terms:
        return set()

    doc_sets = [
        set(inverted_index[term].keys()) for term in terms if term in inverted_index
    ]

    if not doc_sets:
        return set()

    if operator == "AND":
        return (
            set.intersection(*doc_sets) if len(doc_sets) == len(terms) else set()
        )
    else:
        return set.union(*doc_sets)


print("Movies containing both 'space' AND 'alien':", len(boolean_retrieval("space alien", "AND")))
print("Movies containing either 'space' OR 'alien':", len(boolean_retrieval("space alien", "OR")))
```

TF-IDF Query Vector Ranking Function

```python
def search_engine(query, operator="OR"):
    """Processes queries, scores matches via TF-IDF cosine similarity, and formats output."""
    query_terms = preprocess_text(query)
    matched_docs = boolean_retrieval(query, operator=operator)

    if not matched_docs:
        return []

    query_tf = {}
    for term in query_terms:
        query_tf[term] = query_tf.get(term, 0) + 1

    scores = defaultdict(float)

    for term in query_terms:
        if term in inverted_index:
            df_t = len(inverted_index[term])
            idf = math.log(N / (1 + df_t))
            q_weight = query_tf[term] * idf

            for doc_id in matched_docs:
                if doc_id in inverted_index[term]:
                    d_tf = inverted_index[term][doc_id]
                    d_weight = d_tf * idf
                    # Score accumulator
                    scores[doc_id] += q_weight * d_weight

    ranked_results = []
    for doc_id, score in scores.items():
        normalized_score = score / doc_lengths[doc_id]

        movie_row = df[df["id"] == doc_id].iloc[0]
        raw_overview = movie_row["overview"]
        snippet = (
            raw_overview[:140] + "..." if len(raw_overview) > 140 else raw_overview
        ) 

        ranked_results.append(
            {
                "id": doc_id,
                "title": movie_row["title"],
                "snippet": snippet,
                "score": round(normalized_score, 4),
            }
        )

    return sorted(ranked_results, key=lambda x: x["score"], reverse=True)
```

```python
test_query = "christopher nolan space time"
search_results = search_engine(test_query, operator="OR")

print(f"=== Search results for: '{test_query}' ===\n")
for idx, movie in enumerate(search_results[:5], 1):
    print(f"{idx}. {movie['title']} (Match Score: {movie['score']})")
    print(f"   Snippet: {movie['snippet']}")
    print("-" * 50)
```
